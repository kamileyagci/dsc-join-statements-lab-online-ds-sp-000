## Connecting to the Database
Import the necessary packages and connect to the database `'data.sqlite'`.

```python
# Your code here
import pandas as pd
import sqlite3
conn = sqlite3.connect('data.sqlite')
cur = conn.cursor()
```

## Display the names of all the employees in Boston 

Hint: join the employees and offices tables.

```python
# Your code here
cur.execute("""SELECT lastname, firstname
                FROM employees
                JOIN offices
                USING(officeCode)
                WHERE city == 'Boston'
                ;""")

df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df.head()
```

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by.

```python
# Your code here
cur.execute("""SELECT officeCode, city
                FROM offices o
                LEFT JOIN employees o
                USING(officeCode)
                WHERE o.lastname is NULL
                ;""")

df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df.head()
```

## Level Up: Display the names of every individual product that each employee has sold

```python
# Your code here
cur.execute("""SELECT employeeNumber, firstName, lastName, productName, orderNumber, quantityOrdered
                FROM employees e
                JOIN customers c
                ON e.employeeNumber = c.salesRepEmployeeNumber
                JOIN orders o
                USING(customerNumber)
                JOIN orderdetails od
                USING(orderNumber)
                JOIN products p
                USING(productCode)
                ;""")

df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df.head(10)
```

## Level Up: Display the Number of Products each employee has sold

```python
# Your code here
cur.execute("""SELECT employeeNumber, firstName, lastName, sum(quantityOrdered) AS totalQuantitySold
                FROM employees e
                JOIN customers c
                ON e.employeeNumber = c.salesRepEmployeeNumber
                JOIN orders o
                USING(customerNumber)
                JOIN orderdetails od
                USING(orderNumber)
                JOIN products p
                USING(productCode)
                GROUP BY employeeNumber
                ;""")

df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df.head(10)
```
